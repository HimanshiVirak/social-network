### Degree distribution

- Counting the number of edges each node has, known as the **degree sequence** of the network, and looking at the frequency with which each occurs, we can generate a **degree distribution**.

- The sum of the degree sequence will be twice the total number of edges.

- For some classes of complex networks, we often see a degree distribution with a heavy tail.
   

```python
## PLOT FOR Degree Distribution
import matplotlib.pyplot as plt
plt.loglog(range(1,len(ddist)+1),ddist,'o')

plt.title('Social Network')
plt.ylabel('Frequency')
plt.xlabel('Degree')

#SOURCE-CSMAD21_L08_REGRESSION_AND_NETWORKS
```

### Clustering Coefficient

- We can also look at how interconnected the neighbours of each node are. 


- This is known as the **clustering coefficient** of a node [0,1], and we can average this over all nodes to compare networks.


- The local clustering coefficient of a node is defined as:
$$ C = \frac{2E_N}{k(k-1)} $$


- Here, $E_N$ is the total number of edges between neighbours of the node, and $k$ is the number of neighbours.

```python

#Clustering Coefficient

ccg = nx.clustering(SN)
ccg

## Visualise the distribution of cluster coefficients

plt.hist(list(ccg.values()),bins='auto')
plt.title('The distribution of cluster coefficients')
plt.ylabel("Frequency")
plt.xlabel("Clustering coeff.")

#SOURCE-CSMAD21_L08_REGRESSION_AND_NETWORKS
```

#### Description-
**The clustering coefficient of a node can range from 0 to 1. A node with a clustering coefficient of 0 has no triangles involving it, while a node with a clustering coefficient of 1 is part of a fully connected subgraph (a complete graph).**

The clustering function calculates the clustering coefficient for each node in the graph and returns a dictionary with the node labels as keys and the clustering coefficient values as values. The resulting dictionary is stored in the variable ccg.

**It creates a histogram of the values in ccg using Matplotlib. The histogram shows the frequency of nodes with a given clustering coefficient. The x-axis represents the clustering coefficient values and the y-axis represents the frequency of nodes with that clustering coefficient.**

### Betweenness Centrality

- Betweenness centrality is a measure of the **importance** of a node in the network in terms of the possible paths through the network. 

- It is calculated as the *fraction of all shortest paths* in the network that pass through a node.

- Nodes with a high betweenness centrality often connect separate regions of the network, and so are generally important in the function of the network.


```python
# Betweeneness Centrality
bnc = nx.betweenness_centrality(SN)

dict(sorted(bnc.items(), key=lambda kv: kv[1], reverse=True))

#SOURCE-CSMAD21_L08_REGRESSION_AND_NETWORKS
```

```python
#Import 
import matplotlib.pyplot as plt
node_size = [
    v * 50 for v in bnc.values()
]  # set up nodes size for a nice graph representation
plt.figure(figsize=(15, 8))
nx.draw_networkx(SN, node_size=node_size, with_labels=False, width=0.15)
plt.axis("off")
```

#### Description-
**It calculates the betweenness centrality for each node in the graph SN. Betweenness centrality is a measure of how central a node is in the graph, based on the number of shortest paths that pass through it. It is defined as the fraction of all possible pairs of nodes that pass through a given node.**

The betweenness centrality of a node can range from 0 to 1. A node with a betweenness centrality of 0 is not on any shortest paths, while a node with a betweenness centrality of 1 is on all shortest paths.

### Visualising Networks

- To visualise networks we need to assign each node a coordinate in two dimensions. 

- There are many different ways of doing this, and the appropriate method to use depends on the structure of the network.

- One way of doing this is to apply a **force directed layout**. 
    - For each node in the network, nodes are moved based on forces applied by the other nodes of the network.
    - A node is **attracted** towards neighbouring nodes in the graph, but it is **repelled** by other nodes. 
    - The overall force to be applied to the node can then be calculated, and it is moved.
    - This is done over many iterations, allowing large movements initially and then restricting the magnitude of the distance moved by the node.
   

```python
nx.draw_networkx(SN)
```

#### Description-
**Degree centrality assigns a score based on the number of links held by each node — the higher the degree of a node, the more important it is in a graph. Indeed, the more connected is a node, the most information it holds and the most individuals he can quickly connect with.**

```python
G=SN
layout = nx.spring_layout(G)
nx.draw(G,pos=layout,node_size=150,alpha=0.5)

#SOURCE-CSMAD21_L08_REGRESSION_AND_NETWORKS
```

#### Description-
**Identifying all the shortest paths between all pairs of nodes in a graph, and counting how many times a node acts as a bridge along the shortest path between two other nodes, measures betweenness centrality of a node. Nodes with high betweenness may have considerable influence within a network, by virtue of their control over information passing between other nodes.**

```python
G=SN
layout = nx.spectral_layout(G)
nx.draw(G,pos=layout,node_size=150,alpha=0.5)

#SOURCE-CSMAD21_L08_REGRESSION_AND_NETWORKS
```

#### Description-
**The spectral_layout function is used to position the nodes of the graph in two-dimensional space. The draw function is used to draw the graph G using the positions calculated by spectral_layout. The pos argument specifies the positions of the nodes in the graph, node_size specifies the size of the nodes, and alpha specifies the transparency of the nodes.**

```python
G=SN
layout = nx.kamada_kawai_layout(G)
nx.draw(G,pos=layout,node_size=150,alpha=0.5)

#SOURCE-CSMAD21_L08_REGRESSION_AND_NETWORKS
```

#### Description-
**The kamada_kawai_layout function is used for positioning the nodes of a graph in two-dimensional space so that the distances between nodes are proportional to the weights of the edges between them. It can be used to create visually appealing node layouts, especially for undirected graphs.**

```python
G=SN
layout = nx.circular_layout(G)
nx.draw(G,pos=layout,node_size=150,alpha=0.5)

#SOURCE-CSMAD21_L08_REGRESSION_AND_NETWORKS
```

#### Description-
**The circular_layout function positions the nodes of the graph G on a circle. It can be used to create visually appealing node layouts, especially for directed graphs where the edges have a natural direction.**

```python
G=SN
layout = nx.spiral_layout(G)
nx.draw(G,pos=layout,node_size=150,alpha=0.5)

#SOURCE-CSMAD21_L08_REGRESSION_AND_NETWORKS
```

#### Description-
**The spiral_layout function positions the nodes of the graph G in a spiral pattern, starting from the center of the spiral and moving outward. It can be used to create visually appealing node layouts, especially for directed graphs where the edges have a natural direction.**

It simulates a force-directed representation of the network by treating nodes as objects that resist one another and edges as springs that hold them near. This is frequently referred to as an anti-gravity force. Until the positions are somewhat close to equilibrium, simulation is continued.
